```python
import pandas as pd
import numpy as np
```

```python
# ASHRAE
ashrae_train = pd.read_csv("train.csv")
ashrae_meta = pd.read_csv("building_metadata.csv")
ashrae_weather = pd.read_csv("weather_train.csv")

# Building Data Genome
bdg_electricity = pd.read_csv("electricity_cleaned.csv")
bdg_meta = pd.read_csv("metadata.csv")
bdg_weather = pd.read_csv("weather.csv")
```

```python
# Make safe working copies
ashrae_train_prep = ashrae_train.copy()
ashrae_meta_prep = ashrae_meta.copy()
ashrae_weather_prep = ashrae_weather.copy()

bdg_electricity_prep = bdg_electricity.copy()
bdg_meta_prep = bdg_meta.copy()
bdg_weather_prep = bdg_weather.copy()
```

```python
# =========================
# ASHRAE PREPARATION
# =========================

ashrae_train_prep["timestamp"] = pd.to_datetime(ashrae_train_prep["timestamp"], errors="coerce")
ashrae_weather_prep["timestamp"] = pd.to_datetime(ashrae_weather_prep["timestamp"], errors="coerce")

# Drop bad timestamps
ashrae_train_prep = ashrae_train_prep.dropna(subset=["timestamp"])
ashrae_weather_prep = ashrae_weather_prep.dropna(subset=["timestamp"])

# Sort
ashrae_train_prep = ashrae_train_prep.sort_values(["building_id", "timestamp"]).reset_index(drop=True)
ashrae_weather_prep = ashrae_weather_prep.sort_values(["site_id", "timestamp"]).reset_index(drop=True)

print("ASHRAE train after datetime parsing:", ashrae_train_prep.shape)
print("ASHRAE weather after datetime parsing:", ashrae_weather_prep.shape)
```

```python
# Remove duplicate rows
ashrae_train_prep = ashrae_train_prep.drop_duplicates(subset=["building_id", "meter", "timestamp"])
ashrae_weather_prep = ashrae_weather_prep.drop_duplicates(subset=["site_id", "timestamp"])

print("ASHRAE train after duplicate removal:", ashrae_train_prep.shape)
print("ASHRAE weather after duplicate removal:", ashrae_weather_prep.shape)
```

```python
# Replace infinities with NaN
ashrae_train_prep = ashrae_train_prep.replace([np.inf, -np.inf], np.nan)
ashrae_weather_prep = ashrae_weather_prep.replace([np.inf, -np.inf], np.nan)

# Meter reading: impossible negative values
ashrae_train_prep.loc[ashrae_train_prep["meter_reading"] < 0, "meter_reading"] = np.nan

# Weather columns: mark suspicious values as NaN
weather_cols_possible = [
    "air_temperature",
    "cloud_coverage",
    "dew_temperature",
    "precip_depth_1_hr",
    "sea_level_pressure",
    "wind_direction",
    "wind_speed"
]

for col in weather_cols_possible:
    if col in ashrae_weather_prep.columns:
        # Generic clipping by quantiles for extreme outliers
        q_low = ashrae_weather_prep[col].quantile(0.001)
        q_high = ashrae_weather_prep[col].quantile(0.999)
        ashrae_weather_prep.loc[
            (ashrae_weather_prep[col] < q_low) | (ashrae_weather_prep[col] > q_high),
            col
        ] = np.nan

# Extra sanity rules if columns exist
if "cloud_coverage" in ashrae_weather_prep.columns:
    ashrae_weather_prep.loc[
        (ashrae_weather_prep["cloud_coverage"] < 0) | (ashrae_weather_prep["cloud_coverage"] > 10),
        "cloud_coverage"
    ] = np.nan

if "wind_direction" in ashrae_weather_prep.columns:
    ashrae_weather_prep.loc[
        (ashrae_weather_prep["wind_direction"] < 0) | (ashrae_weather_prep["wind_direction"] > 360),
        "wind_direction"
    ] = np.nan

if "wind_speed" in ashrae_weather_prep.columns:
    ashrae_weather_prep.loc[ashrae_weather_prep["wind_speed"] < 0, "wind_speed"] = np.nan

print("ASHRAHE invalid values handled.")
```

```python
# Time-based interpolation inside each site_id
ashrae_weather_prep = ashrae_weather_prep.set_index("timestamp")

numeric_weather_cols = ashrae_weather_prep.select_dtypes(include=[np.number]).columns.tolist()

ashrae_weather_prep[numeric_weather_cols] = (
    ashrae_weather_prep
    .groupby("site_id")[numeric_weather_cols]
    .transform(lambda x: x.interpolate(method="time", limit_direction="both"))
)

ashrae_weather_prep = ashrae_weather_prep.reset_index()

print("ASHRAE weather interpolation done.")
```

```python
ashrae_df = ashrae_train_prep.merge(
    ashrae_meta_prep,
    on="building_id",
    how="left"
)

ashrae_df = ashrae_df.merge(
    ashrae_weather_prep,
    on=["site_id", "timestamp"],
    how="left"
)

print("ASHRAE merged shape:", ashrae_df.shape)
display(ashrae_df.head())
```

```python
# Time features
ashrae_df["hour"] = ashrae_df["timestamp"].dt.hour
ashrae_df["dayofweek"] = ashrae_df["timestamp"].dt.dayofweek
ashrae_df["month"] = ashrae_df["timestamp"].dt.month
ashrae_df["day"] = ashrae_df["timestamp"].dt.day
ashrae_df["is_weekend"] = ashrae_df["dayofweek"].isin([5, 6]).astype(int)

# Create zone-like information from building use if available
if "primary_use" in ashrae_df.columns:
    def map_zone(x):
        x = str(x).lower()
        if any(k in x for k in ["office", "education", "lodging", "retail", "entertainment", "public", "service"]):
            return "commercial"
        elif any(k in x for k in ["manufact", "industrial", "technology", "utility", "warehouse"]):
            return "industrial"
        elif any(k in x for k in ["residential", "apart", "dorm", "housing"]):
            return "residential"
        else:
            return "commercial"
    
    ashrae_df["zone_type"] = ashrae_df["primary_use"].apply(map_zone)

print("ASHRAE feature engineering done.")
display(ashrae_df.head())
```

```python
# =========================
# BDG PREPARATION
# =========================

bdg_electricity_prep["timestamp"] = pd.to_datetime(bdg_electricity_prep["timestamp"], errors="coerce")
bdg_weather_prep["timestamp"] = pd.to_datetime(bdg_weather_prep["timestamp"], errors="coerce")

bdg_electricity_prep = bdg_electricity_prep.dropna(subset=["timestamp"])
bdg_weather_prep = bdg_weather_prep.dropna(subset=["timestamp"])

# Sort
bdg_electricity_prep = bdg_electricity_prep.sort_values("timestamp").reset_index(drop=True)
bdg_weather_prep = bdg_weather_prep.sort_values("timestamp").reset_index(drop=True)

# Convert wide electricity data to long format
bdg_long = bdg_electricity_prep.melt(
    id_vars="timestamp",
    var_name="building_id",
    value_name="meter_reading"
)

print("BDG long shape:", bdg_long.shape)
display(bdg_long.head())
```

```python
bdg_long = bdg_long.replace([np.inf, -np.inf], np.nan)
bdg_weather_prep = bdg_weather_prep.replace([np.inf, -np.inf], np.nan)

# Impossible negative electricity usage
bdg_long.loc[bdg_long["meter_reading"] < 0, "meter_reading"] = np.nan

# Drop duplicate timestamp-building pairs
bdg_long = bdg_long.drop_duplicates(subset=["timestamp", "building_id"])

print("BDG invalid values handled.")
```

```python
# Rename metadata ID column if needed
if "uid" in bdg_meta_prep.columns:
    bdg_meta_prep = bdg_meta_prep.rename(columns={"uid": "building_id"})

bdg_df = bdg_long.merge(
    bdg_meta_prep,
    on="building_id",
    how="left"
)

print("BDG after metadata merge:", bdg_df.shape)
display(bdg_df.head())
```

```python
# Inspect weather columns first
print("BDG weather columns:", bdg_weather_prep.columns.tolist())
display(bdg_weather_prep.head())
```

```python
# Try to merge weather safely only if a common site column exists
common_cols = set(bdg_df.columns).intersection(set(bdg_weather_prep.columns))
print("Common columns between BDG data and weather:", common_cols)

# Typical case: if both have site_id
if "site_id" in bdg_df.columns and "site_id" in bdg_weather_prep.columns:
    bdg_df = bdg_df.merge(
        bdg_weather_prep,
        on=["site_id", "timestamp"],
        how="left"
    )
    print("BDG weather merged using site_id + timestamp.")
else:
    print("BDG weather not merged yet. We will inspect and adapt if needed.")
```

```python
bdg_df["hour"] = bdg_df["timestamp"].dt.hour
bdg_df["dayofweek"] = bdg_df["timestamp"].dt.dayofweek
bdg_df["month"] = bdg_df["timestamp"].dt.month
bdg_df["day"] = bdg_df["timestamp"].dt.day
bdg_df["is_weekend"] = bdg_df["dayofweek"].isin([5, 6]).astype(int)

# Create zone_type from primary use if available
possible_use_cols = ["primaryspaceusage", "primary_use", "industry"]
use_col = None
for c in possible_use_cols:
    if c in bdg_df.columns:
        use_col = c
        break

if use_col is not None:
    def map_zone_bdg(x):
        x = str(x).lower()
        if any(k in x for k in ["office", "education", "commercial", "retail", "service"]):
            return "commercial"
        elif any(k in x for k in ["industrial", "manufact", "factory", "utility", "laboratory"]):
            return "industrial"
        elif any(k in x for k in ["residential", "housing", "dorm", "apartment"]):
            return "residential"
        else:
            return "commercial"
    
    bdg_df["zone_type"] = bdg_df[use_col].apply(map_zone_bdg)

print("BDG feature engineering done.")
display(bdg_df.head())
```

```python
def add_training_features(df):
    df = df.copy()
    df = df.sort_values(["building_id", "timestamp"])

    # time features
    df["hour"] = df["timestamp"].dt.hour
    df["dayofyear"] = df["timestamp"].dt.dayofyear
    df["dayofweek"] = df["timestamp"].dt.dayofweek
    df["month"] = df["timestamp"].dt.month

    # cyclical features
    df["hour_sin"] = np.sin(2 * np.pi * df["hour"] / 24)
    df["hour_cos"] = np.cos(2 * np.pi * df["hour"] / 24)
    df["doy_sin"] = np.sin(2 * np.pi * df["dayofyear"] / 365)
    df["doy_cos"] = np.cos(2 * np.pi * df["dayofyear"] / 365)

    # rolling and difference features
    df["target_roll_mean_24"] = (
        df.groupby("building_id")["meter_reading"]
        .transform(lambda x: x.rolling(24).mean())
    )
    df["target_roll_std_24"] = (
        df.groupby("building_id")["meter_reading"]
        .transform(lambda x: x.rolling(24).std())
    )
    df["target_diff_1h"] = (
        df.groupby("building_id")["meter_reading"]
        .transform(lambda x: x.diff(1))
    )

    # optional transform
    df["target_log"] = np.log1p(df["meter_reading"])

    return df

ashrae_ready = add_training_features(ashrae_ready)
bdg_ready = add_training_features(bdg_ready)

print("ASHRAE after feature engineering:", ashrae_ready.shape)
print("BDG after feature engineering:", bdg_ready.shape)
```

```python
ashrae_ready = ashrae_ready.dropna().reset_index(drop=True)
bdg_ready = bdg_ready.dropna().reset_index(drop=True)

print("ASHRAE after dropna:", ashrae_ready.shape)
print("BDG after dropna:", bdg_ready.shape)
```

```python
def time_based_split(df):
    df = df.sort_values("timestamp").reset_index(drop=True)
    n = len(df)

    train_end = int(n * 0.8)
    val_end = int(n * 0.9)

    train_df = df.iloc[:train_end].copy()
    val_df = df.iloc[train_end:val_end].copy()
    test_df = df.iloc[val_end:].copy()

    return train_df, val_df, test_df

ashrae_train_df, ashrae_val_df, ashrae_test_df = time_based_split(ashrae_ready)
bdg_train_df, bdg_val_df, bdg_test_df = time_based_split(bdg_ready)

print("ASHRAE splits:")
print("Train:", ashrae_train_df.shape)
print("Validation:", ashrae_val_df.shape)
print("Test:", ashrae_test_df.shape)

print("\nBDG splits:")
print("Train:", bdg_train_df.shape)
print("Validation:", bdg_val_df.shape)
print("Test:", bdg_test_df.shape)
```

```python
from sklearn.preprocessing import StandardScaler

ashrae_feature_cols = [
    "target_log", "hour_sin", "hour_cos", "doy_sin", "doy_cos",
    "target_roll_mean_24", "target_roll_std_24", "target_diff_1h"
]

# add weather features if they exist
ashrae_weather_cols = [c for c in [
    "air_temperature", "cloud_coverage", "dew_temperature",
    "precip_depth_1_hr", "sea_level_pressure", "wind_direction", "wind_speed"
] if c in ashrae_train_df.columns]

ashrae_feature_cols = ashrae_feature_cols + ashrae_weather_cols

bdg_feature_cols = [
    "target_log", "hour_sin", "hour_cos", "doy_sin", "doy_cos",
    "target_roll_mean_24", "target_roll_std_24", "target_diff_1h"
]

ashrae_scaler = StandardScaler()
bdg_scaler = StandardScaler()

ashrae_train_df[ashrae_feature_cols] = ashrae_scaler.fit_transform(ashrae_train_df[ashrae_feature_cols])
ashrae_val_df[ashrae_feature_cols] = ashrae_scaler.transform(ashrae_val_df[ashrae_feature_cols])
ashrae_test_df[ashrae_feature_cols] = ashrae_scaler.transform(ashrae_test_df[ashrae_feature_cols])

bdg_train_df[bdg_feature_cols] = bdg_scaler.fit_transform(bdg_train_df[bdg_feature_cols])
bdg_val_df[bdg_feature_cols] = bdg_scaler.transform(bdg_val_df[bdg_feature_cols])
bdg_test_df[bdg_feature_cols] = bdg_scaler.transform(bdg_test_df[bdg_feature_cols])

print("ASHRAE feature columns:", ashrae_feature_cols)
print("BDG feature columns:", bdg_feature_cols)
```

```python
# Timestamp continuity check
def timestamp_continuity_check(df, name):
    df = df.sort_values(["building_id", "timestamp"]).copy()
    gap_stats = (
        df.groupby("building_id")["timestamp"]
        .diff()
        .dropna()
        .dt.total_seconds()
        .div(3600)
    )

    print(f"\n{name} timestamp continuity check")
    print("Min gap (hours):", gap_stats.min())
    print("Median gap (hours):", gap_stats.median())
    print("Max gap (hours):", gap_stats.max())
    print("Non-1h gaps count:", (gap_stats != 1).sum())

timestamp_continuity_check(ashrae_ready, "ASHRAE")
timestamp_continuity_check(bdg_ready, "BDG")
```

```python
# Invalid value check
def invalid_value_check(df, name):
    print(f"\n{name} invalid value check")
    print("Missing meter_reading:", df["meter_reading"].isnull().sum())
    print("Negative meter_reading:", (df["meter_reading"] < 0).sum())
    print("Infinite meter_reading:", np.isinf(df["meter_reading"]).sum())

invalid_value_check(ashrae_ready, "ASHRAE")
invalid_value_check(bdg_ready, "BDG")
```

```python
# Split date-range check
def split_range_check(train_df, val_df, test_df, name):
    print(f"\n{name} split date-range check")
    print("Train:      ", train_df["timestamp"].min(), " --> ", train_df["timestamp"].max())
    print("Validation: ", val_df["timestamp"].min(),   " --> ", val_df["timestamp"].max())
    print("Test:       ", test_df["timestamp"].min(),  " --> ", test_df["timestamp"].max())

split_range_check(ashrae_train_df, ashrae_val_df, ashrae_test_df, "ASHRAE")
split_range_check(bdg_train_df, bdg_val_df, bdg_test_df, "BDG")
```

```python
# Feature scaling / normalization check
def scaling_check(train_df, feature_cols, name):
    print(f"\n{name} scaling check")
    summary = pd.DataFrame({
        "mean_after_scaling": train_df[feature_cols].mean(),
        "std_after_scaling": train_df[feature_cols].std()
    })
    display(summary)

scaling_check(ashrae_train_df, ashrae_feature_cols, "ASHRAE")
scaling_check(bdg_train_df, bdg_feature_cols, "BDG")
```

```python
# Optional sequence-window creation check for LSTM input shape
def create_lstm_sequences(df, feature_cols, target_col="meter_reading", seq_len=24):
    X, y = [], []

    values_X = df[feature_cols].values
    values_y = df[target_col].values

    for i in range(len(df) - seq_len):
        X.append(values_X[i:i+seq_len])
        y.append(values_y[i+seq_len])

    return np.array(X), np.array(y)

# Example on ASHRAE train set
ashrae_X_train, ashrae_y_train = create_lstm_sequences(
    ashrae_train_df,
    ashrae_feature_cols,
    target_col="meter_reading",
    seq_len=24
)

print("ASHRAE LSTM X shape:", ashrae_X_train.shape)
print("ASHRAE LSTM y shape:", ashrae_y_train.shape)
```
